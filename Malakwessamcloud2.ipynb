```python
import pandas as pd
df=pd.read_csv("popularbooks.csv")
df
```

```python
#check for null
df.info()
```

```python
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
df
```

```python
#drop one of book id coloumn becuase they are the same
df = df.drop(columns=['goodreads_book_id'])
#find average rate of ratings from one to 5 and replace it with the new one
df['average_rating'] = df[['ratings_1','ratings_2','ratings_3', 'ratings_4','ratings_5']].mean(axis=1)
df.drop(columns=['ratings_1','ratings_2','ratings_3','ratings_4','ratings_5'], inplace=True)
df
```

```python
#filter data for Harry Potter series
harry_potter_books = df[df['original_title'].str.contains('Harry Potter', case=False)]

# Clean the data (remove any rows with missing values)
harry_potter_books.dropna(inplace=True)
df
```
